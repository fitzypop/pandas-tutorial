```python
import pandas as pd
import numpy as np

people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}

df = pd.DataFrame(people)
# Changing 'bad' data to nan
# None will be captured by dropna, so it doesn't neec to be changed
df.replace('NA', np.nan, inplace=True)
df.replace('Missing', np.nan, inplace=True)
df
```

```python
df.dropna()
```

```python
# only drop rows with all nan values
df.dropna(axis='index', how='all')
```

```python
# only drop rows with all nan values
df.dropna(axis='columns')
```

```python
# only drop rows with all nan values
df.dropna(subset=['email'])
```

```python
# only drop a row if both lastname and email are missing
df.dropna(how='all', subset=['last', 'email'])
```

```python
df.isna()
```

```python
# Change na values, best for numerical data
df.fillna(0)
```

```python
df['age'] = df['age'].astype(float)
df.dtypes
```
